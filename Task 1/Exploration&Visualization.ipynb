# Importing Liabraries, Loading Dataset and Printing Shape

```python
# Import the pandas library for data manipulation
import pandas as pd

# Import seaborn library for data visualization
import seaborn as sns

# Import matplotlib's pyplot and rename it as 'pt' for plotting graphs
import matplotlib.pyplot as pt

# Import numpy library for numerical operations
import numpy as np

# This magic command is used in Jupyter notebooks to display plots inline (inside the notebook)
%matplotlib inline

# Define the file path of the Iris dataset CSV file
file_path = "Iris.csv"

# Read the CSV file into a pandas DataFrame, using the 'Id' column as the index
iris_data = pd.read_csv(file_path, index_col="Id")

# Display the number of rows and columns in the dataset
iris_data.shape
```

# Scatter Plot

```python

# Set the figure size for the plot (width=12 inches, height=6 inches)
pt.figure(figsize=(12, 6))

# Create a scatter plot using the iris_data DataFrame
# (Note: You should specify 'x' and 'y' axes explicitly for a meaningful plot)
sns.scatterplot(
    data=iris_data
)

# Set the label for the y-axis
pt.ylabel("Size in cm")

# Display the plot
pt.show()
```

#  Insights from the above Scatter Plot of the Whole Data and the features:
The length of the sepals for all the species is the greatest than the width of sepals and petals as well of petals length.
The length of petals for most of the species is greater than the width of petals and sepals and smaller than of sepals length. While a few species of petals has smaller length sepals width, petals width and sepal length.
A few species of petals has the smallest width of seapals width, petals length and sepals length. Although most of the species has wider petals  and some are equal in the size of petals length


```python
# Set the size of the plot to 12 inches wide and 6 inches tall
pt.figure(figsize=(12, 6))

# Create a scatter plot using seaborn
# x-axis: PetalWidthCm, y-axis: PetalLengthCm, hue: color-coded by Species
sns.scatterplot(
    data=iris_data,
    x="PetalWidthCm", 
    y="PetalLengthCm", 
    hue="Species"
)

# Display the plot
pt.show()
```

#  Insights from the above Scatter Plot of the Petal Length and Petal Width:
There are 3 species of Iris Flowers:
Iris Setosa
Iris versicolor
Iris verginica
Iris Setosa has smaller petals in length(1cm-1.9cm) and width(0.1cm-0.6cm) as compared to Iris versicolor and Iris verginica.
Iris versicolor has larger petals in length (3cm-4.9cm) and width(1cm-1.5) than of Iris Setosa but smaller petals in length and width than of Iris verginica. And a very few has wider petals than a negligible number of Iris verginica.
Iris verginica has the largest petals length and width of other 2 species of Iris.

```python
# Set the size of the figure to 12 inches wide and 6 inches tall
pt.figure(figsize=(12, 6))

# Create a scatter plot with:
# x-axis: SepalWidthCm
# y-axis: SepalLengthCm
# hue: color-coded by Species (different colors for each flower type)
sns.scatterplot(
    data=iris_data,
    x="SepalWidthCm", 
    y="SepalLengthCm", 
    hue="Species"
)

# Display the plot
pt.show()
```

#  Insights from the above Scatter Plot of the Sepal Length and Sepal Width:
The sepal length(4.9cm-7cm) and width(2cm-3.5cm) of Iris Versicolor and verginica specie varies.
While the sepals of iris setosa are the widest from both of the two species but have small petals length


```python
# Set the size of the plot to 12 inches wide and 6 inches tall
pt.figure(figsize=(12, 6))

# Create a scatter plot with:
# x-axis: SepalWidthCm (width of the sepal)
# y-axis: PetalWidthCm (width of the petal)
# hue: Use different colors for each iris species
sns.scatterplot(
    data=iris_data,
    x="SepalWidthCm", 
    y="PetalWidthCm", 
    hue="Species"
)

# Display the plot
pt.show()
```

The iris sitosa has the highest sepal width and the iris verginica has the highest petal width.

```python
# Set the size of the plot to 12 inches wide and 6 inches tall
pt.figure(figsize=(12, 6))

# Create a scatter plot with:
# x-axis: PetalLengthCm (length of the petal)
# y-axis: SepalLengthCm (length of the sepal)
# hue: Different colors for each species of iris
sns.scatterplot(
    data=iris_data,
    x="PetalLengthCm", 
    y="SepalLengthCm", 
    hue="Species"
)

# Display the plot
pt.show()
```
